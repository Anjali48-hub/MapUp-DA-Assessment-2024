```python
#question 1
def reverse_in_groups(lst, n):
    result = []
    for i in range(0, len(lst), n):
        group = []

        for j in range(min(n, len(lst) - i)):
            group.insert(0, lst[i + j])
        result.extend(group)
    return result
```

```python
#question 2

def group_by_length(strings):
    length_dict = {}


    for string in strings:
        length = len(string)
        if length not in length_dict:
            length_dict[length] = []
        length_dict[length].append(string)


    return dict(sorted(length_dict.items()))
```

```python
#question 3

def flatten_dict(d, parent_key='', separator='.'):
    items = {}

    for key, value in d.items():
        new_key = f"{parent_key}{separator}{key}" if parent_key else key

        if isinstance(value, dict):
            # recursively flatten dictionaries
            items.update(flatten_dict(value, new_key, separator))

        elif isinstance(value, list):
            # flatten lists by adding index as part of the key
            for i, item in enumerate(value):
                list_key = f"{new_key}[{i}]"
                if isinstance(item, dict):
                    items.update(flatten_dict(item, list_key, separator))
                else:
                    items[list_key] = item

        else:
            # for non-dictionary, non-list items, just assign the value
            items[new_key] = value

    return items
```

```python
#question 4:
def unique_permutations(nums):
    def backtrack(path, remaining, result):
        if not remaining:
            result.append(path)
            return

        seen = set()
        for i in range(len(remaining)):
            if remaining[i] not in seen:
                seen.add(remaining[i])
                backtrack(path + [remaining[i]], remaining[:i] + remaining[i+1:], result)

    result = []
    backtrack([], sorted(nums), result)
    return result
```

```python
#question5

import re

def find_all_dates(text):
    # Regular expression patterns for different date formats
    date_patterns = [
        r'\b\d{2}-\d{2}-\d{4}\b',    # dd-mm-yyyy
        r'\b\d{2}/\d{2}/\d{4}\b',    # mm/dd/yyyy
        r'\b\d{4}\.\d{2}\.\d{2}\b'   # yyyy.mm.dd
    ]

    # Compile the patterns into a single regex
    date_regex = '|'.join(date_patterns)

    # Find all matches of dates in the text
    matches = re.findall(date_regex, text)

    return matches
```

```python
#question7

def rotate_and_transform(matrix):
    n = len(matrix)

    # Step 1: Rotate the matrix by 90 degrees clockwise
    rotated_matrix = [[0] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            rotated_matrix[j][n - 1 - i] = matrix[i][j]

    # Step 2: Transform the rotated matrix
    final_matrix = [[0] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            row_sum = sum(rotated_matrix[i])  # sum of elements in the same row
            col_sum = sum(rotated_matrix[k][j] for k in range(n))  # sum of elements in the same column
            final_matrix[i][j] = row_sum + col_sum - 2 * rotated_matrix[i][j]  # exclude the current element

    return final_matrix
```

```python
#question 8

import pandas as pd

def check_timestamps_completeness(df):
    # Define the full range of days and times
    full_days = {'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'}
    full_times = pd.date_range('00:00:00', '23:59:59', freq='S').time

    # Function to check if a single (id, id_2) pair covers all days and times
    def is_complete(group):
        days_covered = set()
        times_covered = set()

        for _, row in group.iterrows():
            start_day = row['startDay']
            end_day = row['endDay']
            start_time = pd.to_datetime(row['startTime']).time()
            end_time = pd.to_datetime(row['endTime']).time()

            # Add all days in the range
            days_covered.update(pd.date_range(start=start_day, end=end_day, freq='D').strftime('%A'))

            # Add all times in the range
            if start_day == end_day:
                times_covered.update(pd.date_range(start=start_time, end=end_time, freq='S').time)
            else:
                times_covered.update(full_times)

        return days_covered == full_days and len(times_covered) == len(full_times)

    # Group by (id, id_2) and apply the completeness check
    completeness = df.groupby(['id', 'id_2']).apply(is_complete)

    return completeness

```

```python
#question 9

import pandas as pd
import numpy as np

def calculate_distance_matrix(file_path):
    # Read the dataset
    df = pd.read_csv("/content/dataset-2.csv")

    # Get unique IDs
    ids = pd.unique(df[['id_start', 'id_end']].values.ravel('K'))

    # Initialize the distance matrix with infinity
    distance_matrix = pd.DataFrame(np.inf, index=ids, columns=ids)

    # Set diagonal to 0
    np.fill_diagonal(distance_matrix.values, 0)

    # Fill the matrix with direct distances
    for _, row in df.iterrows():
        distance_matrix.at[row['id_start'], row['id_end']] = row['distance']
        distance_matrix.at[row['id_end'], row['id_start']] = row['distance']

    # Use the Floyd-Warshall algorithm to compute shortest paths
    for k in ids:
        for i in ids:
            for j in ids:
                if distance_matrix.at[i, j] > distance_matrix.at[i, k] + distance_matrix.at[k, j]:
                    distance_matrix.at[i, j] = distance_matrix.at[i, k] + distance_matrix.at[k, j]

    return distance_matrix



```

```python
#question 10

import pandas as pd

def unroll_distance_matrix(distance_matrix):
    # Initialize an empty list to store the rows
    rows = []

    # Iterate over the distance matrix to unroll it
    for id_start in distance_matrix.index:
        for id_end in distance_matrix.columns:
            if id_start != id_end:
                distance = distance_matrix.at[id_start, id_end]
                rows.append({'id_start': id_start, 'id_end': id_end, 'distance': distance})

    # Create a DataFrame from the list of rows
    unrolled_df = pd.DataFrame(rows)

    return unrolled_df

```

```python
#question 11

import pandas as pd

def find_ids_within_ten_percentage_threshold(df, reference_id):
    # Calculate the average distance for the reference_id
    reference_distances = df[df['id_start'] == reference_id]['distance']
    if reference_distances.empty:
        return []

    reference_avg_distance = reference_distances.mean()

    # Calculate the 10% threshold
    lower_bound = reference_avg_distance * 0.9
    upper_bound = reference_avg_distance * 1.1

    # Find ids within the 10% threshold
    ids_within_threshold = df.groupby('id_start')['distance'].mean()
    ids_within_threshold = ids_within_threshold[(ids_within_threshold >= lower_bound) & (ids_within_threshold <= upper_bound)].index.tolist()

    # Sort the list of ids
    ids_within_threshold.sort()

    return ids_within_threshold

```

```python
#question 12

import pandas as pd

def calculate_toll_rate(df):
    # Define rate coefficients
    rate_coefficients = {
        'moto': 0.8,
        'car': 1.2,
        'rv': 1.5,
        'bus': 2.2,
        'truck': 3.6
    }

    # Calculate toll rates for each vehicle type
    for vehicle, rate in rate_coefficients.items():
        df[vehicle] = df['distance'] * rate

    return df

```

```python
#question 13

import pandas as pd
import numpy as np
from datetime import time

def calculate_time_based_toll_rates(df):
    # Define discount factors
    weekday_factors = {
        '00:00:00-10:00:00': 0.8,
        '10:00:00-18:00:00': 1.2,
        '18:00:00-23:59:59': 0.8
    }
    weekend_factor = 0.7

    # Define days of the week
    days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

    # Initialize lists to store new columns
    start_days = []
    start_times = []
    end_days = []
    end_times = []

    # Initialize columns for toll rates
    df['moto'] = df['distance'] * 0.8
    df['car'] = df['distance'] * 1.2
    df['rv'] = df['distance'] * 1.5
    df['bus'] = df['distance'] * 2.2
    df['truck'] = df['distance'] * 3.6

    # Iterate over each row to calculate time-based toll rates
    for _, row in df.iterrows():
        for day in days_of_week:
            if day in ['Saturday', 'Sunday']:
                factor = weekend_factor
                start_days.append(day)
                end_days.append(day)
                start_times.append(time(0, 0, 0))
                end_times.append(time(23, 59, 59))
                df.loc[_, 'moto'] *= factor
                df.loc[_, 'car'] *= factor
                df.loc[_, 'rv'] *= factor
                df.loc[_, 'bus'] *= factor
                df.loc[_, 'truck'] *= factor
            else:
                for time_range, factor in weekday_factors.items():
                    start_time_str, end_time_str = time_range.split('-')
                    start_time_obj = pd.to_datetime(start_time_str).time()
                    end_time_obj = pd.to_datetime(end_time_str).time()
                    start_days.append(day)
                    end_days.append(day)
                    start_times.append(start_time_obj)
                    end_times.append(end_time_obj)
                    df.loc[_, 'moto'] *= factor
                    df.loc[_, 'car'] *= factor
                    df.loc[_, 'rv'] *= factor
                    df.loc[_, 'bus'] *= factor
                    df.loc[_, 'truck'] *= factor

    # Add new columns to the DataFrame
    df['start_day'] = start_days
    df['start_time'] = start_times
    df['end_day'] = end_days
    df['end_time'] = end_times

    return df

```
